# **Arithmetic Operations**

```python
a = np.arange(1,10).reshape(3,3)
b = np.arange(11,20).reshape(3,3)
print(a)
print(b)
print("SUM: ",a + b)
print("SUB: ",a - b)
print("MULTIPICATION: ",a * b)
print("DIV: ",b / a)
print("FLOOR DIVISION: ",b // a)
print("MODULO: ",b % a)
```

```python
a = np.arange(1,10).reshape(3,3)
np.subtract(a,3) # Subtract
```

```python
a = np.arange(1,10).reshape(3,3)
np.sum(a) # Sum
```

```python
a = np.array([(1,1,2,3,4,5), (1,3,5,4,67,7), (1,2,8,4,6,7)])
np.sum(a, axis=0) # Column-wise sum
```

# **Condition Apply**

```python
a = np.array([(1,1,2,3,4,5), (1,3,5,4,67,7), (1,2,8,4,6,7)])
a[a < 50] = 0 # replace value
print(a)
```

```python
a = np.array([1, 2, 3, 4, 5])

result = np.where(a % 2 == 0, a * 10, a)
print(result)
```

# **Split Array**

```python
a = np.arange(1,37).reshape(6,6)
print(a)
np.hsplit(a,3) # horizontal split
```

```python
a = np.arange(1,37).reshape(6,6)
print(a)
np.split(a,3, axis=1) # horizontal split
```

```python
a = np.arange(1,37).reshape(6,6)
print(a)
np.split(a,2, axis=0) # vertical split
```

```python
a = np.arange(1,37).reshape(6,6)
print(a)
np.vsplit(a,3) # vertical split
```

```python
a = np.arange(1,37).reshape(6,6)
np.array_split(a, 4, axis =1) #unequal horizontal split
```

```python
a = np.arange(1,37).reshape(6,6)
np.array_split(a, 4, axis =0) #unequal vertical split
```

# **Stack Array**

```python
a = np.array([(1,1,2,3,4,5), (1,3,5,4,67,7), (1,2,8,4,6,7)])
b = np.array([(1,34,10,3,64,5), (1,33,25,42,67,7)])

np.vstack((a,b)) # vertical stack array
```

```python
a = np.array([(1,1,2,3,4,5), (1,3,5,4,67,7), (1,2,8,4,6,7)])
b = np.array([(1,34,10,3,64,5), (1,33,25,42,67,7), (11,23,38,4,65,7)])

np.hstack((a,b)) # horizontal stack array
```

```python
a = np.array([(1,1,2,3,4,5), (1,3,5,4,67,7), (1,2,8,4,6,7),(1,2,8,4,6,7)])
np.clip(a,5,100) # it is doing like the conver the all lower number to this perticular number
```

```python
a = np.array([(1,1,2,3,4,5), (1,3,5,4,67,7), (1,2,8,4,6,7),(1,2,8,4,6,7)])
np.where(a > 7, 1, 0) # where condition
```

# **Mathematics Calculation**

```python
a = np.array([(1,1,2,3,4,5), (1,3,5,4,67,7), (1,2,8,4,6,7)])
b = np.array([(1,34,10),(3,64,5), (1,33,25),(42,67,7), (11,23,38),(4,65,7)])

np.dot(a,b) #dot array \(C_{i,j}=\sum _{k=0}^{n-1}a_{i,k}\cdot b_{k,j}\)
```

```python
a = np.array([(1,1,2,3,4,5), (1,3,5,4,67,7), (1,2,8,4,6,7)])
np.std(a) # Standard deviation
```

```python
a = np.array([(1,1,2,3,4,5), (1,3,5,4,67,7), (1,2,8,4,6,7)])
np.var(a) # Variance
```

```python
a = np.array([(1,1,2,3,4,5), (1,3,5,4,67,7), (1,2,8,4,6,7),(1,2,8,4,6,7)])
np.argmax(a) # index of max
```

```python
a = np.array([(1,1,2,3,4,5), (1,3,5,4,67,7), (1,2,8,4,6,7),(1,2,8,4,6,7)])
np.argmin(a) # index of min
```

```python
a = np.array([(1,1,2,3,4,5), (1,3,5,4,67,7), (1,2,8,4,6,7),(1,2,8,4,6,7)])
b =(a - a.min()) / (a.max() - a.min())
print(b) # normalization
```

```python
a = np.array([(1,1,2,3,4,5), (1,3,5,4,67,7), (1,2,8,4,6,7)])
b = np.array([(1,34,10),(3,64,5), (1,33,25),(42,67,7), (11,23,38),(4,65,7)])
np.matmul(a,b) # same as dot function (multiplication)
```

```python
a = np.array([[ 196,  782,  273],
       [ 948, 2655, 2773],
       [ 277, 1287,  525]])
np.linalg.eig(a) #Eigen values
```

```python
a = np.array([[ 196,  782,  273],
       [ 948, 2655, 2773],
       [ 277, 1287,  525]])
np.linalg.inv(a) #Inverse matrix
```

```python
a = np.array([[ 196,  782,  273],
       [ 948, 2655, 2773],
       [ 277, 1287,  525]])
np.linalg.det(a) #Determinate
```

```python
a = np.array([[ 196,  782,  273],
       [ 948, 2655, 2773],
       [ 277, 1287,  525]])
np.percentile(a,87.5) #Percentile
```

```python
a = np.array([(1,1,2,3,4,5), (1,3,5,4,67,7), (1,2,8,4,6,7)])
b = np.array([(1,34,10,3,64,5), (1,33,25,42,67,7), (11,23,38,4,65,7)])
np.corrcoef(a,b) #Correlation
```

```python
a = np.array([(1,1,2,3,4,5), (1,3,5,4,67,7), (1,2,8,4,6,7)])
b = np.array([(1,34,10,3,64,5), (1,33,25,42,67,7), (11,23,38,4,65,7)])
np.cov(a,b) #Covariance
```
